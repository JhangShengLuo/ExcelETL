```python
# ---------設定字的格式
def set_style(name,height,bold=False):
    import xlwt
    style = xlwt.XFStyle() # 初始化样式
    font = xlwt.Font() # 为样式创建字体                    
    font.name = name # 'Times New Roman'
    font.bold = bold
    font.color_index = 4
    font.height = height
    # borders= xlwt.Borders()
    # borders.left= 6
    # borders.right= 6
    # borders.top= 6
    # borders.bottom= 6
    style.font = font  # style.borders = borders
    return style
# ---------
```

```python
#讀取內湖資料只抓出設定之內湖區各時段分開的街道資料，將三個時段資料存成三個資料夾（上/下/全日）。
# -*- coding: utf-8 -*-
def writePeriod09():
    import os
    if not os.path.exists('上午時段'):
        os.makedirs('上午時段')
    if not os.path.exists('下午時段'):
        os.makedirs('下午時段')
    if not os.path.exists('全日時段'):
        os.makedirs('全日時段')
    
    
    files = []

    for f in os.listdir('.'):
            if os.path.isfile(f):
                    files.append(f)
    for f in files:
        if '日報表' in f:
            import xlrd,xlwt
            wbk = xlwt.Workbook()
            data = xlrd.open_workbook(f)
            sheet = wbk.add_sheet('sheet1',cell_overwrite_ok=True)
            table = data.sheets()[0]

            j=0
            for i in range(0,57,+3):#只取上午時段

                rows = table.row_values(i)
                j+=1
                for ele in range(0,len(rows)):
                    sheet.write(j-1,ele,rows[ele],set_style('Times News Romans',200,True))



            sheet.write(j,5,"=SUM(F1:F19)")


            # print len(rows)
            # for ele in rows:
            #     print ele
            # sheet.write(0,1,table)
            wbk.save('上午時段/'+'上午'+f)
        
    
            
#     下午時段        
    files = []
    for f in os.listdir('.'):
            if os.path.isfile(f):
                    files.append(f)
    for f in files:
        if '日報表' in f:
            import xlrd,xlwt
            wbk = xlwt.Workbook()
            data = xlrd.open_workbook(f)
            sheet = wbk.add_sheet('sheet1',cell_overwrite_ok=True)
            table = data.sheets()[0]

            j=0
            for i in range(1,57,+3):#只取上午時段

                rows = table.row_values(i)
                j+=1
                for ele in range(0,len(rows)):
                    sheet.write(j-1,ele,rows[ele],set_style('Times News Romans',200,True))



            sheet.write(j,5,"=SUM(F1:F19)")


            # print len(rows)
            # for ele in rows:
            #     print ele
            # sheet.write(0,1,table)
            wbk.save('下午時段/'+'下午'+f)
        
            


#      全日時段        
    files = []
    for f in os.listdir('.'):
            if os.path.isfile(f):
                    files.append(f)
    for f in files:
        if '日報表' in f:
            import xlrd,xlwt
            wbk = xlwt.Workbook()
            data = xlrd.open_workbook(f)
            sheet = wbk.add_sheet('sheet1',cell_overwrite_ok=True)
            table = data.sheets()[0]

            j=0
            for i in range(2,57,+3):#只取上午時段

                rows = table.row_values(i)
                j+=1
                for ele in range(0,len(rows)):
                    sheet.write(j-1,ele,rows[ele],set_style('Times News Romans',200,True))



            sheet.write(j,5,"=SUM(F1:F19)")


            # print len(rows)
            # for ele in rows:
            #     print ele
            # sheet.write(0,1,table)
            wbk.save('全日時段/'+'全日'+f)
        
```

```python
#  把抓出來的內湖區的資料，取檔案日期跟當日所有路段的當量做總和，在寫成一個csv檔

def periodVolume09():
    # -*- coding: utf-8 -*-
    import os,re
    import csv
#     上午

    try:
        
        newcsv = open('上午時段/上午.csv','w')
        files = []
        for f in os.listdir('./上午時段'):
            files.append(f)
        for f in files:
            if '日報表' in f:
                import xlrd,xlwt
                wbk = xlwt.Workbook()
                data = xlrd.open_workbook('上午時段/'+f)
                table = data.sheets()[0]
                cols = table.col_values(5)
                sumall = 0
                for i in range(0,len(cols)-1):
                #print cols[i]
                    sumall += float(cols[i])
                match = re.findall('[0-9]+.[0-9]+.[0-9]+',f)
    #             print str(match[0]) , str(sumall) # 解開註解可以印出來看 
                newcsv.write(str(match[0])+','+str(sumall)+'\n')
        newcsv.close()
        
    except:
        print 'no 上午時段 directory'
    try:

    #     下午
        newcsv = open('下午時段/下午.csv','w')
        files = []
        for f in os.listdir('./下午時段'):
            files.append(f)
        for f in files:
            if '日報表' in f:
                import xlrd,xlwt
                wbk = xlwt.Workbook()
                data = xlrd.open_workbook('下午時段/'+f)
                table = data.sheets()[0]
                cols = table.col_values(5)
                sumall = 0
                for i in range(0,len(cols)-1):
                #print cols[i]
                    sumall += float(cols[i])
                match = re.findall('[0-9]+.[0-9]+.[0-9]+',f)
    #             print str(match[0]) , str(sumall) # 解開註解可以印出來看 
                newcsv.write(str(match[0])+','+str(sumall)+'\n')
        newcsv.close()
    except:
        print 'no 下午時段 directory'
    try:

    #     全日

        newcsv = open('全日時段/全日.csv','w')
        files = []
        for f in os.listdir('./全日時段'):
            files.append(f)
        for f in files:
            if '日報表' in f:
                import xlrd,xlwt
                wbk = xlwt.Workbook()
                data = xlrd.open_workbook('全日時段/'+f)
                table = data.sheets()[0]
                cols = table.col_values(5)
                sumall = 0
                for i in range(0,len(cols)-1):
                #print cols[i]
                    sumall += float(cols[i])
                match = re.findall('[0-9]+.[0-9]+.[0-9]+',f)
    #             print str(match[0]) , str(sumall) # 解開註解可以印出來看 
                newcsv.write(str(match[0])+','+str(sumall)+'\n')
        newcsv.close()
        
    except:
        print 'no 全日時段 directory'
```

```python
def morningAndNoon():

    import io
    import csv
    if not os.path.exists('上下午'):
        os.makedirs('上下午')

    morning = open('上午時段/上午.csv','rU')
    noon = open('下午時段/下午.csv','rU')
    twoPeriod = open('上下午/上下午.csv','w')

    # 上午
    mrlist = []
    mr = csv.reader(morning,delimiter=' ')
    for mrow in mr:
        mrlist.append(mrow)


    # 下午
    nrlist = []
    nr = csv.reader(noon,delimiter=' ')
    for nrow in nr:
        nrlist.append(nrow)

    # print len(nrlist)


    for i in range(0,len(mrlist)):
    #     print nrlist[i][0]
        twoPeriod.write(mrlist[i][0]+','+'am')
        twoPeriod.write('\n')

        twoPeriod.write(nrlist[i][0]+','+'pm')
        twoPeriod.write('\n')



    morning.close()
    noon.close()
    twoPeriod.close()
```
